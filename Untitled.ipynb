# EDA 2 - Post Cleaned

## Count Unique

```python
# Count group by City-State
count_city_state = df.groupby('State')['City'].nunique().reset_index()
count_city_state = count_city_state.sort_values(by = 'City', ascending = False)
count_city_state
```

```python
# Count group by Make-Model
count_make_model_nunqie = df.groupby('Make')['Model'].nunique().reset_index()
count_make_model_nunqie = count_make_model_nunqie.sort_values(by = 'Model', ascending = False)
count_make_model_nunqie
```

## Count Sales

```python
# Count sales brand
count_make_model = df.groupby('Make')['Model'].count().reset_index()
count_make_model = count_make_model.sort_values(by = 'Model', ascending = False)
count_make_model
```

```python
# Count state - model
count_state_model = df.groupby('State')['Model'].count().reset_index()
count_state_model = count_state_model.sort_values(by = 'Model', ascending = False)
count_state_model
```

```python
# Group by 'State' and 'Make', and count the sales
model_popularity1 = df.groupby(['State', 'Make']).size().reset_index(name='Count')

# Sort within each state by sales count in descending order
model_popularity2 = model_popularity1.sort_values(by=['State', 'Count'], ascending=[True, False])

# Get the most popular model for each state
popular_makes = model_popularity2.groupby('State').first()
popular_makes = popular_makes.sort_values('Count', ascending = False)

popular_makes
```

```python
# Lower casing
df['State'] = df['State'].str.upper()
df['City'] = df['City'].str.lower()
df['Make'] = df['Make'].str.lower()
df['Model'] = df['Model'].str.lower()

df
```
