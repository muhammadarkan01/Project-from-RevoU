# EDA

```python
import pandas as pd
```

```python
df = pd.read_csv("true_car_listings.csv")
df
```

```python
# Count city, state, make, model

city_val = df['City'].nunique()
state_val = df['State'].nunique()
make_val = df['Make'].nunique()
model_val = df['Model'].nunique()

print("City: ", city_val)
print("State: ", state_val)
print("Make: ", make_val)
print("Model: ", model_val)
```

```python
# Value in each city, state, make, model

city_val = df['City'].unique()
state_val = df['State'].unique()
make_val = df['Make'].unique()
model_val = df['Model'].unique()

print("City: ", city_val)
print("State: ", state_val)
print("Make: ", make_val)
print("Model: ", model_val)

```

## Count

```python
# Count group by City-State
count_city_state = df.groupby('State')['City'].nunique().reset_index()
count_city_state
```

```python
# Count group by Make-Model
count_make_model_nunqie = df.groupby('Make')['Model'].nunique().reset_index()
count_make_model_nunqie
```

```python
# Count sales brand
count_make_model = df.groupby('Make')['Model'].count().reset_index()
count_make_model
```
